### Supervised ML Algos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits, load_iris
from sklearn.svm  import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
```

```python
RANDOM_STATE = 24
np.random.seed(RANDOM_STATE)
```

```python
iris = load_iris()
iris.keys()
```

```python
print(iris.DESCR)
```

```python
iris.data.shape, iris.target.shape
```

```python
x = iris.data
y = iris.target
x.shape, y.shape
```

```python
plt.figure(figsize=(15,8))

for count in range(x.shape[1]):
    plt.subplot(2,2,count+1)
    n,bins,patches = plt.hist(x[:,count],100)
    plt.title(iris.feature_names[count])

plt.tight_layout()
```

```python
from sklearn.naive_bayes import GaussianNB

clf = GaussianNB()
clf.fit(x,y)

y_pred = clf.predict(x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)
```

```python
xplot = x[:,0]
yplot = x[:,1]

x_err = xplot[(y != y_pred)]
y_err = yplot[(y != y_pred)]
x_err.shape, y_err.shape
```

```python
target_names = iris.target_names
formatter = plt.FuncFormatter(lambda i, *arg:target_names[int(i)])
```

```python
plt.scatter(xplot,yplot, s=10, c=y, cmap=plt.cm.brg)
plt.colorbar(ticks=[0,1,2], format=formatter)
plt.scatter(x_err, y_err, s=50, color='w', edgecolors='black', alpha = 0.5)
```
